My first workbook where I will show a simple example on how to fetch and display stock data

## Getting the data

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt 

ticker = 'DANSKE.CO'
```

```python
data = yf.download(ticker, period = '1y')
```

```python
import matplotlib.pyplot as plt 
```

```python
data['Close'].plot(grid = True)
```

Calculate daily returns:

```python
returns = data['Close'].pct_change()
```

```python
returns.plot()
```

```python
returns.hist(bins=50, figsize=(12,8))
```

```python
import numpy as np
startdate = np.datetime64('2020-04-01')
print(startdate)
```

```python
returns.loc[startdate:].head()
```

```python
cum_daily_return = (1 + returns.loc[startdate:]).cumprod()
```

```python
cum_daily_return.plot()
```
